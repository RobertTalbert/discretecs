```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Random 8-node graph

G = nx.gnp_random_graph(8, 0.5)

nx.to_dict_of_lists(G)
nx.draw_circular(G, node_color='cyan', with_labels=True)
plt.show()
```

```python
# Make a table of distances between each node

nx.floyd_warshall_numpy(G)
```

```python
# Find the eccentricity of each node

nx.eccentricity(G)
```

```python
# Find the diameter and radius of G

print(nx.diameter(G))
print(nx.radius(G))
```
