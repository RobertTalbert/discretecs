```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.gnp_random_graph(8, 0.35)
nx.draw(g, with_labels=True, node_color = "lightgreen", pos=nx.circular_layout(g))
```

```python
nx.is_eulerian(g)
```

```python
# Does the complete graph K_n ever have an Euler circuit?
n = 4
h = nx.complete_graph(n)
nx.draw(h, with_labels=True, node_color = "lightgreen", pos=nx.circular_layout(h))
```

```python
nx.is_eulerian(h)
```

```python
nx.eulerian_circuit(h)
```

```python
[x for x in nx.eulerian_circuit(h)]
```
