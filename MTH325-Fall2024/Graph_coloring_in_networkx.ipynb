```python
# Load the libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Generate a graph graph to work with, and visualize
g = nx.gnp_random_graph(16, 0.3)
nx.draw(g, with_labels=True, pos = nx.spring_layout(g),
        node_color = "lightgray")
plt.show()
```

```python
# Vertex colorings in Python are given as dictionaries.
# Use the "coloring" method in networkX.
# There are several different algorithms available, greedy being one of them.

coloring = nx.coloring.greedy_color(g)
print(coloring)
```

```python
# The default strategy for ordering the vertices is "largest first", meaning
# it goes in descending order of vertex degrees.

# You can change the initial vertex ordering in the greedy algorithm by
# passing an additional parameter "stategy" to greedy_color.
# For example this one uses a randomized order:

random_coloring = nx.coloring.greedy_color(g, strategy="random_sequential")
print(random_coloring)
```

[Full documentation on the greedy algorithm in networkX here](https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.coloring.greedy_color.html).

Note, there is no method in networkX that produces a guaranteed minimal coloring, and no method in networkX that computes the chromatic number of a graph. There are only coloring algorithms and strategies, none of which is guaranteed to produce a minimal vertex coloring. In general, finding a minimal coloring of a graph is an [NP-complete problem](https://en.wikipedia.org/wiki/NP-completeness).

```python
# To actually color the vertices of a graph with a particular coloring:

# Extract the colors assigned to each vertex with a list comprehension
colors = [coloring[node] for node in g.nodes()]

# Redraw the graph
# The first line is unchanged from above except for the node colors
# THe second line is where we apply the colors to the nodes
nx.draw(g, with_labels=True, pos = nx.spring_layout(g),
        node_color=colors, cmap=plt.cm.rainbow)
```
