## Motivating example: Social media

Social media services like Twitter, Facebook, Instagram, and so on can be thought of as relations on a set of people. Specifically, for example, let $F$ be the set of all user accounts on Facebook. Define the "Facebook relation" to be the relation on $F$, in which a pair of users `(user1, user2)` is an element of the relation if `user1` has "friended" `user2`. So for example, if Alice sends a friend request to Bob on Facebook, and Bob accepts, then `(Alice, Bob)` becomes a pair in the relation. Facebook itself can be thought of as an immense set of ordered pairs, one pair for every pair of people who friend each other. 

Twitter works this way too. If $T$ is the set of all user accounts on Twitter, define the "Twitter relation" to be the relation on $T$ defined so that a pair of users `(user1, user2)` is an element of the relation if `user1` has "followed" `user2`. For example, on Twitter, I (`@RobertTalbert`) follow the account for the Liverpool FC soccer team (`@LFC`). So `(@RobertTalbert, @LFC)` is an element of the relation. 

But if you think about these two social media services, they are quite different in at least one major way: On Facebook, you cannot friend another person without that other person accepting your friend request, whereas on Twitter it's quite common for a user to follow many other accounts without being followed back. For example, on Facebook if Alice friended Bob, essentially Bob friended Alice too. In terms of ordered pairs, if `(Alice, Bob)` is an element of the relation, then `(Bob, Alice)` must also be an element of the relation or else the "friendship" is one-way. The relationship is inherently "symmetric". On the other hand, Twitter is highly non-symmetric; for example I follow \@LFC but Liverpool FC does not follow me, sadly. In terms of of ordered pairs, `(@RobertTalbert, @LFC)` is an element of the relation but `(@LFC, @RobertTalbert)` is not. 

So "symmetry" is a **property** of these social media relations; Facebook "has symmetry" whereas Twitter does not. The difference in this property leads to important differences in how these social media services behave. Facebook tends to feel more like a group of friends due to its symmetry; whereas Twitter is less intimate and more attuned to just watching what people say. Both approaches to social media have pros and cons, but the point is that the difference in look and feel arises from a difference in a _mathematical property_ about the set of ordered pairs that defines the social networks. In this lesson, we want to study four such properties that are highly descriptive of social and other networks. 

## Four recurring examples 

To illustrate the four properties of relations in this lesson, we'll introduce four relations, all defined on the set $U = \{1,2,3,\dots, 10\}$: 

+ **The "divides" relation:** A pair $(a,b)$ is an element of this relation if $a$ divides $b$. For example, $(2,10)$ is in this relation but $(4,10)$ is not. 
+ **The "next to" relation:** A pair $(a,b)$ is an element of this relation if $a$ is directly adjacent to $b$ on the number line. A somewhat fancier way of saying this is that $(a,b)$ is in the relation if $|a-b| = 1$. For example $(5,4)$ is in this relation but $(2,4)$ is not. 
+ **The "less than or equal to" relation:** A pair $(a,b)$ is an element of this relation if $a \leq b$. For example $(3,10)$ is in this relation, and so is $(3,3)$. But $(6,3)$ is not. 
+ **The "strict less than" relation:** A pair $(a,b)$ is an element of this relation if $a < b$. For example $(3,10)$ is in this relationBut $(6,3)$ is not, and neither is $(3,3)$.

These may not be as interesting as Facebook and Twitter, but they are simple and easy to play with. In fact we can generate the list of pairs in each one quite easily using list comprehensions in Python: 

```python
U = range(1,11)  # The "set" {1,2,..., 10}

# Divides:
divides = [(a,b) for a in U for b in U if b % a == 0]
print(divides)
```

```python
# Next to:
next_to = [(a,b) for a in U for b in U if abs(a-b) == 1]
print(next_to)
```

```python
# Less than/equal to
less_than_eq = [(a,b) for a in U for b in U if a <= b]
print(less_than_eq)
```

```python
# Less than/equal to
strict_less_than = [(a,b) for a in U for b in U if a < b]
print(strict_less_than)
```

 ## Reflexive relations
 
 
The `divides` and `less_than_eq` relations have a certain property in common that `next_to` and `strict_less_than` do not have. Look closely at the set (list) of pairs in those relations. Do you see it? 

Answer: In `divides` and `less_than_eq`, **every element of $U$ is related to itself.** In the language of ordered pairs, $(a,a)$ is an element of the relation (same first and second coordinate) for every $a \in U$. In the other two relations, this isn't the case. 

This motivates the formal definition of the first of our properties: 

>__Definition:__ Let $r$ be a relation on the set $A$. Then $r$ is said to be __reflexive__ if $(a,a) \in r$ for every $a \in A$. 

In a reflexive relation, every point in the set on which the relation is defined is related to itself. In a non-reflexive relation, there may be _some_ points that are related to themselves but there is at least one point that is not related to itself. So, `divides` and `less_than_eq` are reflexive relations; `next_to` and `strict_less_than` are not reflexive. 

**NOTE:** The opposite of "every point is related to itself" is not "none of the points are related to themselves". Therefore a non-reflexive relation would _not_ be one in which there are _no_ pairs of the form $(a,a)$. A non-reflexive relation would be one in which **there exists a point $a \in U$** such that $(a,a)$ is not in the relation. For example, here is a non-reflexive relation on $U$ in which some points are related to themselves, but some aren't: 

>{(0,0), (1,1), (2,2), (2,3), (3,4)} 

This is a relation on $U$ because it's a subset of $U \times U$. Some of the points of $U$ (0,1,2) are related to themselves; others (3,4,5,...,10) aren't. That makes this a non-reflexive relation. In other words:

>The relation $r$ on $A$ is __not reflexive__ if _there exists_ at least one $a \in U$ such that $(a,a) \not \in r$. 

Going back to social networks: Social networks are very rarely reflexive. A reflexive social network would be one in which a user can follow, friend, or connect with himself. Aside from being a little creepy and egotistical, friending or following oneself is usually not allowed by the rules of the network. (More precisely, an _account_ cannot connect with itself. If I made two different Twitter accounts both for myself and had them follow each other, this wouldn't count as reflexive.)  

What do reflexive relations _look like_? Let's play with some representations of our four examples and use networkX to help. 


```python
import warnings
warnings.filterwarnings('ignore')
import networkx as nx
import matplotlib.pyplot as plt

# Converting the list of pairs to actual networkX digraphs: 

divides_relation = nx.DiGraph(divides)
next_to_relation = nx.DiGraph(next_to)
leq_relation = nx.DiGraph(less_than_eq)
slt_relation = nx.DiGraph(strict_less_than)

# Put these into a list: 
examples = [divides_relation, next_to_relation, leq_relation, slt_relation]

# Look at the dictionaries: 
for r in examples:
    print(nx.to_dict_of_lists(r))
```

Looking at the dictionaries for these relations, it becomes easy to spot the reflexive ones (first and third ones above): Every key contains itself in its list of values. For the non-reflexive relations (second and fourth ones), this isn't the case. 

What about digraphs? Let's look at the one for divides: 

```python
nx.draw_circular(divides_relation, with_labels=True)
plt.show()
```

This is actually a little misleading. In this relation, we know that $(1,1)$ is an element so **there should be an edge from 1 to itself.** This would be a directed self-loop. We haven't drawn these so far because they are "against the rules" for simple graphs, but for digraphs of relations we _do_ want to include directed self-loops if it's appropriate. Technically, on this digraph, _every node_ should have a self-loop. Unfortunately: **There is no simple way to produce self-loops in networkX graphs.** So we have to leave this up to our own mental images and hand drawings. However, in a digraph for a reflexive relation, every node should have a self-loop; for a non-reflexive relation at least one node will be missing a self-loop. 

## Symmetric relations

We alluded to "symmetry" in the opening examples with Twitter and Facebook by saying that Facebook "has symmetry" whereas Twitter does not. Let's be a little more precise about what we mean with that. 

Look at the digraph for the "next to" relation, paying particular attention to the arrowheads: 

```python
nx.draw_circular(next_to_relation, with_labels=True)
plt.show()
```

Compare this to the digraph for "divides" and you might see a difference in the arrows. Namely: **every edge in the "next to" relation points in both directions** whereas **not every edge in the "divides" relation is two-directional**. In the language of sets of tuples, we see that in the next-to relation, whenever $(a,b)$ belongs to the list, so does $(b,a)$. (Remember that a tuple here is really just an abstracted version of an arrow.)

That gets us to the next definition: 

>__Definition:__ Let $r$ be a relation on the set $A$. Then $r$ is said to be __symmetric__ if whenever $(a,b) \in r$, we also have $(b,a) \in r$. 

Notice that this definition, unlike the one for reflexive, is a _conditional statement_ -- it's in the form of "if-then". _If_ $(a,b) \in r$ _then_ $(b,a) \in r$. It does not say anything about _whether_ two points are connected in the first place. For example, in the next-to relation there is no edge between 1 and 3; but this doesn't mean the relation is not symmetric. 

A relation _fails_ to be symmetric when _there exists_ a pair $(a,b)$ in the relation but $(b,a)$ is not. For example, there is an edge from $3$ to $6$ in the "divides" relation but not from $6$ to $3$, and that's what makes "divides" not symmetric. 

What about the other two relations? Are they symmetric? 

+ Less than or equal to: This is _not symmetric_ because there is a pair in the relation whose "reverse pair" is not, for example $(1,3)$ is in the relation but $(3,1)$ isn't. 
+ Strict less-than: Also not symmetric and for the same reason. 

You can really _see_ the failure of symmetry in these two if you look at the digraphs. Here's the one for strict less-than: 

```python
nx.draw_circular(slt_relation, with_labels=True)
plt.show()
```

You can just look at this digraph and tell that the relation is not symmetric because it's very easy to find an arrow that points one way but not the other. 

## Antisymmetric relations

Just to drive home the point, here's another relation on the same set ($U = \{1,2,\dots,10\}$) that is not symmetric, although there are _some_ edges that are bi-directional:

```python
not_symmetric = [(1,2), (2,1), (1,3), (3,1), (2,4), (4,2), (2,5), (3,6), (4,7), (5,8), (6,9), (7,10)]
not_symmetric_relation = nx.DiGraph(not_symmetric)
nx.draw_circular(not_symmetric_relation, with_labels=True)
plt.show()
```

_Some_ of the edges are two-directional, but **not all of them are** (for example the edge from 2 to 5) and this is why the relation is not symmetric. Twitter can be like this; there are people who I follow who follow me back, and there are people I follow who don't. Being non-symmetric doesn't mean the relation is _devoid_ of symmetry.

But go back to the divides relation and look at its graph again:

```python
nx.draw_circular(divides_relation, with_labels=True)
plt.show()
```

This relation is a more extreme example: _The arrows **NEVER** point both ways._ They are always one-way. This relation really _is_ devoid of symmetry. 

Here is a formal definition that we will use to capture the meaning of this concept: 

>__Definition:__ Let $r$ be a relation on a set $A$. Then $r$ is said to be __antisymmetric__ if for all $a,b \in A$, whenever $(a,b) \in r$ and $(b,a) \in r$, we must have $a = b$. 

This is a standard but somewhat confusing definition, so let's look at it informally first and then give an alternative but logically equivalent definition. 

Informally, this definition is saying that if two "different" points have a two-directional relationship, they really _aren't_ two different points -- they are _equal_ to each other. This is really captured by the "divides" relation: If two positive integers divide each other, then the only logical conclusion is that they are the same integer. 

But it still can be weird to think about two "different" points being the same, so here is an alternative definition that might be clearer. It is using the [contrapositive](http://www.regentsprep.org/regents/math/geometry/gp2/lcontrap.htm) of the if-then statement in the main definition. 

>__Alternate definition:__ Let $r$ be a relation on a set $A$. Then $r$ is said to be __antisymmetric__ if given $a,b \in A$, if $a \neq b$, then either $(a,b) \not \in r$ or $(b,a) \not \in r$. 

In English, this is saying that in an antisymmetric relation, if we have two different points, it must be the case that one of them is not related to the other. **In other words there cannot be two-directional relations _anywhere_ in an antisymmetric relation.** This is exactly what you saw with the "divides" graph -- no two-directional relations. (Sometimes no relations at all -- but never any going both ways.) So the "divides" relation is antisymmetric. 

To _fail_ to be antisymmetric, a relation would have at least one pair of points where there _is_ a two-directional relation -- like in Twitter. Therefore Twitter and other social networks where two-directional relations are allowed are _not_ antisymmetric. This seems to be the case for _all_ social networks; can you think of a social network where two-directional relations are against the rules? (One possible such "social network" that comes to mind is a military or corporate chain of command, where officers or vice presidents report to others "up the chain" but never backwards. This would be a network that _is_ antisymmetric because there are no bi-directional connections.)

Let's check the digraphs for the other three example relations to see if they are antisymmetric: 

```python
nx.draw_circular(next_to_relation, with_labels=True)
plt.show()
```

The next-to relation is _definitely not_ antisymmetric because not only do bi-directional edges exist, every single one of them is bidirectional! 

Here are the less-than-equal-to and strict-less-than relations:

```python
nx.draw_circular(leq_relation, with_labels=True)
plt.show()
```

```python
nx.draw_circular(slt_relation, with_labels=True)
plt.show()
```

These look the same because networkX won't plot self-loops. But the one for less-than-equal-to should have lots of loops. Either way, there are no two-directional edges on either of these graphs; so the relations are both antisymmetric. This makes sense when you think about how "less than or equal to" works. If you know that $a \leq b$ and $b \leq a$, then the only way this can happen is  $a = b$. 

## Transitive relations 

It's time for our last property, and it's considered by those who use relations in computer applications to be the most important one. 

Go back and look at the directed graphs for the "divides" and "next to" relations, which you generated above. For each one, do the following exercise: **Can you find a pair of points that are connected by a path of exactly two edges, that are _not_ directly connected?** Look through several pairs of points in each relation.

What you should find is: 

+ In the "divides" relation, this never happens. Any two points that are connected by a path of two edges, are actually directly connected. 
+ But in the "next to" relation, there are lots of times where there are points connected by a path of length 2 that are not directly connected. For example there is a path of length 2 from 5 to 7, but 5 and 7 are not adjacent. 

So in some relations, like "next to", it's possible for nodes to be "distantly related" without being actually related. While in others, _any_ length of connection, no matter how distant, forces a direct connection. 

This gives us our last definition: 

>__Definition:__ Let $r$ be a relation on a set $A$. Then $r$ is said to be __transitive__ if for all $a,b,c \in A$, if $(a,b) \in r$ and $(b,c) \in r$, we have $(a,c) \in r$. 

This captures the essence of the little exercise you just did. In a transitive relation, if two nodes are connected by a path of two edges ($(a,b)$ and $(b,c)$ are both in the relation) then they must be directly connected ($(a,c)$ is in the relation) and this must happen _every time_ two nodes are connected by a path of length 2. 

A relation would _fail_ to be transitive if there existed two nodes that were connected by a path of length 2 that were _not_ directly connected, like in the "next to" relation. Even if a relation has _some_ instances where such nodes are directly connected but just a _single instance_ where they are not, then the relation fails to be transitive. Note the universal quantifier in the definition. 

Let's check out our recurring examples again for transitivity: 

+ **Divides:** This relation _is transitive_. You can see it on the digraph that we've produced several times now; if $a$ "points to" $b$ and $b$ "points to" $c$, then $a$ "points to" $c$. This is a way of expressing the basic arithmetic theorem that says if $a$ divides $b$ and $b$ divides $c$, then $a$ divides $c$. 
+ **Next to:** This is _not transitive_ as described above. 
+ **Less than or equal to, and strict less than:** These are both transitive because of $a < b$ and $b < c$, then $a < c$ as we know from elementary school. 

In social networks, transitivity often fails. LinkedIn, a social network for people in professional fields, is a good example. On LinkedIn, I follow my friend Todd Stanislav (a professor at Ferris State University) and Todd follows someone named Stephen Ehrmann. But I do not follow Stephen Ehrmann (nor had I ever heard of this person until just now). So, LinkedIn is not transitive. Stephen is not my "friend" but rather a "friend of a friend", or a _second-level connection_. 


## A look ahead: Transitive closure 

LinkedIn isn't transitive, but _it wants to be transitive_. What I mean by this is, I get notifications from LinkedIn every day that say, _"You follow X and X follows Y. Would you like to add Y to your network?"_  LinkedIn is constantly prompting me to "close the loop" on my professional network by adding "friends of friends" directly. Suppose I got a notification asking me to add Stephen Ehrmann because I know Todd Stanislav and Todd knows Stephen. And suppose that I do so, so that I am now no longer two edges away from Stephen but directly connected. Well, now I will start getting notifications asking me to add some of Stephen's friends, who used to be three edges away from me in the network but are now two, and LinkedIn wants to get that to one. And then I'd be prompted to add friends of _those_ people, and on and on it goes. 

If I "closed out" my entire network so that every person to whom I am connected by a _path_ -- not just of two edges but of _any_ length of edges -- were added as direct connections, then I'd stop getting those LinkedIn notifications. I would have what is called the **transitive closure** of my LinkedIn network and this is the ideal state that LinkedIn wants me to have. What would this "ultimate network" look like? And how would I know when I have it? 

"Closing the loop on distant connections" like this is actually a very important idea for _any_ network -- for example a network of fiber optic cables that connect buildings on a campus. If Mackinac is connected to Kindschi and Kindschi is connected to Kirkhof by expensive fiber-optic cables, there are reasons people might want to directly connect Mackinac and Kirkhof. 

Soon, we'll consider an algorithm for finding the transitive closure of a relation that is highly efficient and uses the adjacency matrix for a relation (which we didn't look at in this tutorial).
