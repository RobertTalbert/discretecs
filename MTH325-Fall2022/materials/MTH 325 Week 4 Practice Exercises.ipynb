## Graph representation (Content Skill Standard G.1) 

>★ G.1: I can represent a graph in different ways and change representations from one to another.

Take a look at this graph, and write it as an edge list, an adjacency list, a Python dictionary, and an adjacency matrix. 

```python
import networkx as nx
G = nx.gnp_random_graph(8, 0.4) 
nx.draw_circular(G, with_labels = True, node_color = "white")
```

**You can self-check your work using the following Python/networkX commands:**

```python
# To convert G to an edge list

nx.to_edgelist(G)

# To convert G to an edge list without the empty weights given
list(G.edges())
```

```python
# To convert G to a dictionary

# networkx doesn't really have a way of converting to adjacency lists; 
# the dictionary is basically what this is. 

nx.to_dict_of_lists(G) 
```

```python
# To convert G to an adjacency matrix: 
# This technically creates a "data frame" not a matrix, 
# but it looks like a matrix so we'll use it. 

nx.to_pandas_adjacency(G)
```
