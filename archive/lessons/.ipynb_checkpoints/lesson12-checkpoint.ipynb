## Background

### Partial orderings 

Everybody knows that you can put _numbers_ in order. For example there is a really natural way to order the following numbers: 

> $5$, $\sqrt{5}$, $5 \pi$, $-5$

And that, of course, is using the idea of "less than or equal to". This set of numbers (_any_ set of numbers) comes with a natural way of ordering the elements using the simple $\leq$ concept. (The ordering of this set is $-5, \sqrt{5}, 5, 5\pi$.) 

But people also realize that the concept of _ordering_ isn't just for numbers. For example we put _words_ in order all the time, using alphabetical order. And if you are given a list of words, it's very easy to put them in order using alphabetical ordering into one "chain" of ordered words. 

However, there are some instances where we have sets of objects that can be put in order, but it's a little more complicated than $\leq$ on numbers or alphabetical ordering on words. 

+ Example: Look at the collection of all the subsets of the set $\{1,2,3\}$. This is a collection of eight sets: 
$$\emptyset, \{1\}, \{2\}, \{3\}, \{1,2\}, \{1,3\}, \{2,3\}, \{1,2,3\}$$
There is a way of thinking about these subsets in which one set can be "less than" another. For example you probably agree that $\{1\}$ is "less than" $\{1,2\}$ and $\{1,2\}$ is "less than" $\{1,2,3\}$. But what about $\{1,2\}$ and $\{1,3\}$? Is one of these "less than" the other? Unlike $\leq$ or alphabetical ordering, there is an ordering present but it's not always the case that any two elements are "comparable". 
+ Example: Here is one of my favorite recipes, for [chili-spiced bacon-wrapped asparagus](https://www.evernote.com/l/ABSS22q0m0hJl7Bzwwl3J2BMfpk6689_Jow). Note the "Directions" for the recipe. They are numbered 1--6. Clearly some steps should happen before others -- for example step 3 (cut the bacon into pieces) has to happen before step 4 (wrap the asparagus in the bacon). So we could say that cutting the bacon up is "less than" -- in some sense -- wrapping the asparagus. But not all steps have to go in order. For example, step 1 (soaking the toothpicks) and step 2 (preheating the grill) could be done at the same time. 

So many sets have a sort of natural ordering to them, whether numbers or words or subsets or directions in a recipe. The ordering is not always "in a straight line" so that given any two objects, one must come before the other. But still, there's an ordering present. 

What we'd like to do from a mathematics standpoint is making this concept of "ordering" precise, and yet generic enough to be applied to many different situations. To do this, let's think about ordering as a __relation on the set__. Namely, on the set whose elements we wish to put in order, define a relation by saying __for all $x,y$ in the set, $(x,y)$ belongs to the relation if $x$ "comes before" $y$.__ 

Applied to our examples, this very generic definition would look like this: 

+ Less-than-or-equal-to relation on a set of numbers: Given numbers $x,y$ let $(x,y)$ be in the relation if $x \leq y$. 
+ Alphabetical order on a set of words: Given words $w_1, w_2$ let $(w_1, w_2)$ be in the relation if $w_1$ comes before $w_2$ in alphabetical order, or if $w_1$ and $w_2$ are the same word. For example $(\text{dog, dude})$ would be in the relation but not $(\text{dog, cat})$ 
+ Subsets: Given subsets $A,B$ of the set $\{1,2,3,4\}$, let $(A,B)$ be in the relation if $A \subseteq B$ -- that is, if $A$ is a subset of, or is equal to $B$. For example $(\{1\}, \{1,2\})$ would belong to this relation but not $(\{1,2,3\}, \{1,2\})$ and also not $(\{1,2\}, \{1,3\})$.
+ Recipe: Given steps $x$ and $y$ of the recipe, let $(x,y)$ be in the relation if $x = y$ or if $x$ must be completed in order for $y$ to be completed. For example (step 3, step 4) would be in the relation but not (step 1, step 2). 

Since we have defined this concept of ordering in terms of relations, we should think about what _properties_ these relations have. Just like you did in Lesson 11, think about the four big properties we usually associate with relations and for each of the relations we just discussed, use the table below to check off which properties the relation has. 

| Property | $\leq$ relation? | Alphabetical order relation? | Subset relation? | Recipe relation? | 
|:-------- | :--------------- | :-----------------------  | :------ | :------ |
|Refexive |    |    |   |  |
|Symmetric |    |    | |  |
|Antisymmetric |    |    | |  |
|Transitive |    |    |  | |  |

You did the exercise above, right? NO PEEKING. 

OK -- You should have found that __all of these relations have the reflexive, antisymmetric, and transitive properties__. (We had to put some awkward wording in a couple of these in order to get reflexivity.) 

So all of these somewhat natural concepts of "putting things in order" can be thought of as relations that have the same three properties. That means we can abstract this idea, using this definition:

>__Definition:__ Let $r$ be a relation on a set $A$. Then $r$ is said to be a __partial ordering__ of $A$ if $r$ is reflexive, antisymmetric, and transitive. If $r$ is a partial ordering then $A$ is a __partially ordered set__ or __"poset"__. 

In other words a partial ordering is any relation that "behaves like" the less-than-or-equal-to relation does on sets of numbers. But it's a general enough idea that all kinds of sets and orderings can be included under this umbrella.

Some further examples of partial orderings and posets: 

+ The "divides" relation on the set $\mathbb{P}$ of positive integers (that is, $(a,b)$ is in the relation if $a$ divides $b$) is a partial order. 
+ The relation on the set $\{1,2,3,4\}$ given by this directed graph is a partial order: 

```python
poset = DiGraph([(1,1), (2,2), (3,3), (4,4), (1,2), (2,3), (1,3), (1,4)], loops=true)
poset.show()
```

__Why is it called a "partial" ordering?__, you're asking. It's because in some partial orderings, it may be the case that two objects are not related at all. For example, look at the set $\{1,2,5,8,10\}$ and place the "divides" relation on it. It's easy to see that some points are not related, for example 5 and 8. Hence the word "partial" in partial ordering. 

A partial ordering in which _any two points are related_ is called a _total ordering_ and their applications in computer science are numerous. For example, we can define a form of recursion and induction on a totally ordered set because given any two objects we can talk about "one coming before the other" which is really the only required ingredient in an induction proof. 

### Hasse diagrams

We've seen that visualizing relations using directed graphs can really help us understand the behavior of the relation. If the relation is a partial ordering, we can remove some extraneous data from its directed graph and have a visualization that is simpler but has the same information in it. This simplified graph is called a __Hasse diagram__ ("Hasse" rhymes with "boss" and "T. Haas") and here are the instructions for how to make one. (Followed by an example of using SageMath to make one.) 

Before we begin, we need to have a _finite_ partially ordered set. Let the set be called $S$ and we'll use the special symbol $\preceq$ to denote the partial ordering. (It looks like a $\leq$ but it has more curves.) For example if $(a,b)$ is in the partial order, then we will write $a \preceq b$. 

1. Draw the directed graph for the relation. 
2. Remove all loops. These are not necessary because all partial orderings are reflexive, so the loops don't tell us anything we didn't already know about this relation. 
3. Remove all edges that would be forced to be in the relation because of the presence of other edges and transitivity. For example: In the directed graph above, There is an edge from 1 to 2, and from 2 to 3, and from 1 to 3. We would remove the edge from 1 to 3 because its existence would be implied by the other two edges plus the transitive property. (This process is sometimes called "transitive reduction".) 
4. Finally, rearrange the edges so that for each edge, the initial (starting) node is below its terminal (ending) node if we drew it on paper. Then remove all the arrows and replace them with undirected line segments. (In other words: Instead of using arrows to denote direction, use the up-down orientation of the graph on paper or a screen.) 

Let's do an example, using the set $\{1,2,3,4,6,8,12\}$ and the "divides" relation. 

First we draw the digraph: 

```python
# Too lazy to type in the edges manually, so here's a script that will do it: 
points = [1,2,3,4,6,8,12]
edgelist = [ ]
for a in points: 
    for b in points:
        if b % a == 0:
            edgelist.append((a,b))   # I.e. if a divides b, add (a,b) to the list of edges
            
poset = DiGraph(edgelist, loops=true)
poset.show()
```

First, remove the loops: 

```python
poset.remove_loops()  # There is a method for just about everything. 
                      # Note that this actually changes the graph definition. If we want the loops back, we'll
                      # have to add them back in manually. 
poset.show()
```

Now we have to think about removing edges that whose existence is implied by the presence of other edges and transitivity. For example, we would remove (1,8) because we already have (1,4) and (4,8) -- and (1,8) is implied because the relation is transitive. Then we'd also remove (1,4) because we already have (1,2) and (2,4) -- the edge (1,4) is implied by transitivity. And so on. 

Remember we mentioned that this removal process is called "transitive reduction". 

```python
poset = poset.transitive_reduction()  # Again, a method for almost everything. However note this time that 
                                      # this particular method does NOT change the graph "in place" -- we have to 
                                      # set the reduced graph equal to the output. 
poset.show()
```

So "transitive reduction" removes a lot of the unnecessary data (or "noise") from the directed graph. (You may be tempted to reduce ALL graphs this way, but remember this process is only for partial orderings!) 

The final step is to rearrange the nodes so that instead of using arrows, we use an up-down orientation. In the rearranged graph, node 1 would be literally below nodes 2 and 3; node 2 would be below nodes 4 and 6, but _not_ below node 3 or vice versa because there is no edge from 2 to 3 or vice versa. Node 6 would be below node 12; node 4 would be below nodes 8 and 12; and finally nodes 8 and 12 would not be below anything -- they would be on the top of the new graph. 

You can almost see the final result by tilting your head sideways (or the screen sideways!) and looking at the reduced graph above. The result would look like this: 

<img src="posethd.png">

So the Hasse diagram for a poset takes the most essential information out of the poset and turns it into a minimalist, simplified visualization.

## Appendix: Creating Hasse diagrams in SageMath

SageMath will draw Hasse diagrams, but it's not as easy or straightforward as drawing directed graphs. 

First, we collect the data we need as if we were going to create the directed graph. For example, here is the partially ordered set we used in the Hasse diagram example above, as a dictionary:

```python
p = {1:[1,2,3,4,6,8,12], 2:[2,4,6,8,12], 3:[3,6,12], 4:[4,8,12], 6:[6,12], 8:[8], 12:[12]}
```

The next step is essential: __Remove all loops from the poset.__ We can do this manually (just go through and delete "1" from the list of values for 1, "2" from the list of values for 2, etc.), or write a script to do it, or use the `.remove_loops()` method as follows (which involves converting to a directed graph first, then running the method, then recovering the dictionary: 

```python
pgraph = DiGraph(p)
pgraph.remove_loops()
newp = pgraph.to_dictionary()
```

```python
newp
```

Having removed the loops, we now create a `Poset` object using the data: 

```python
awesome_poset = Poset(newp)
```

Finally, we just call the `.show()` method on the poset: 

```python
awesome_poset.show()
```

SageMath insists on putting arrowheads in the diagram, although these are not necessary and technically should not be there. Oh well. We just live with it. 

Here is a function that you can run that plots a Hasse diagram using the data for a directed graph, that combines all these steps: 

```python
def drawHasse(graph_data):
    directed_graph = DiGraph(graph_data, loops=true)
    directed_graph.remove_loops()
    poset = Poset(directed_graph.to_dictionary())
    poset.show()
```

For example, here's the Hasse diagram for the poset `[(1,1), (2,2), (4,4), (6,6), (1,2), (1,4), (1,6), (2,4), (2,6), (4,4), (6,6)]`:

```python
drawHasse([(1,1), (2,2), (4,4), (6,6), (1,2), (1,4), (1,6), (2,4), (2,6)])
```
