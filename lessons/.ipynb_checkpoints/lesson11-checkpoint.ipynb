## Background

### Equivalence relations

We often work with sets of objects where we want to think of two of them as being "basically the same" without insisting that they should be _literally_ the same. 

Here is an example that my kids used to bring home from kindergarten: They would bring home a list of words, all different, and would be ask to draw a circle using the same color around words that rhymed. For example, if the list consisted of 

>Boat, heat, cat, hat, seat, float, eat

then they should put blue circles around "boat" and "float", red circles around "heat" and "seat" and "eat", and green circles around "cat" and "hat". The idea in kindergarten is to teach kids that although these words all end in "-at", they do not all sound the same; and by grouping the "sounds the same" words together using colors, we are treating them as _essentially the same with respect to how they are pronounced_. That is, we are grouping the words based on some notion of _equivalence_. None of the words are literally equal to each other; but they are "equivalent" in the sense that they have the same ending sound (i.e. they rhyme). 

Here is another example from the world of information systems. You have noticed, surely, that web sites often have different "top level domains" -- these are the ending strings in their URLs and include `.com`, `.net`, `.gov`, `.org` and a few others. (Actually, [over 1000 others](https://en.wikipedia.org/wiki/List_of_Internet_top-level_domains).) These different top-level domains tend to indicate differences in content, while websites that have the same top-level domain tend to be similar in their content and audience. If we were studying websites with a view toward understanding content and audience, we might find it useful to _group together websites based on their top-level domains_ -- and treat two websites as "equivalent" if they have the same top-level domain. (That is, all the `.com` websites would be considered equivalent, all the `.org` websites would be considered equivalent, and so on.) The websites are not _literally the same_, obviously, but we treat "equivalent" ones the same way, just like my kids would pronounce the ending of two words that rhyme the same way. 

You might notice that __these notions of "equivalence" can actually be thought of as relations.__ For example: 

+ Let $W$ be the set of all words in the English language. Put a relation on $W$ as follows: If $w_1$ and $w_2$ are two words, then $(w_1, w_2)$ belongs to the relation if $w_1$ rhymes with $w_2$. 
+ Let $S$ be the set of all websites on the internet. Put a relation on $S$ as follows: If $s_1$ and $s_2$ are two websites, then $(s_1, s_2)$ belongs to the relation if $s_1$ has the same top-level domain as $s_2$. 

So for instance, in the "rhyming" relation, the pair $(\text{cat, hat})$ would belong to the relation but $(\text{cat, seat)}$ would not. In the top-level domain relation, $(\texttt{cnn.com}, \texttt{foxnews.com})$ would belong to the relation but $(\texttt{cnn.com}, \texttt{rtalbert.org})$ would not. 

Stop for a minute, and without reading any further, check off which of the properties each of these relations have. 

| Property | Rhyming relation? | Top-level domain relation? | 
|:-------- | :--------------- | :-----------------------  | 
|Refexive |    |    | 
|Symmetric |    |    | 
|Antisymmetric |    |    | 
|Transitive |    |    |  |


You did the exercise above, right? Without peeking? 

OK -- You should have found that __both of these relations have the reflexive, symmetric, and transitive properties.__ (Neither has the antisymmetric property -- but we are going to focus for now on the properties they _do_ have instead of the one they _don't_.)  

Having these three properties makes the two relations very similar to the relation of literal equality. When we say two things (numbers, sets, whatever) are "equal", it's understood that 

+ Every "thing" is equal to itself (reflexive)
+ If "thing 1" is equal to "thing 2", then "thing 2" is equal to "thing 1" (symmetric)
+ If "thing 1" is equal to "thing 2", and "thing 2" is equal to "thing 3" then "thing 1" is equal to "thing 3" (transitive)

So _literal equality_  has all three of these properties. A relation that also has all three of these properties may not have to do with objects being _literally equal_ but the relation has all the properties that are at the core of what makes things literally equal: reflexivity, symmetry, and transitivity. 

With all this background, the following definition should make sense: 

>__Definition:__ Let $r$ be a relation on a set $A$. We say $r$ is an __equivalence relation__ if $r$ is reflexive, symmetric, and transitive. 

A math professor of mine in college used to call equivalence relations "RST relations" as a way of remembering the three properties that make an equivalence relation what it is. 

Examples: 

+ Let $A$ be any set of numbers, and define a relation on $A$ by saying $(x,y)$ is in the relation ix $x = y$. That is, this relation is _literal equality_. Then as we discussed, this is an equivalence relation. In fact it is the prototypical equivalence relation. 
+ The "rhymes with" relation on a set of words is an equivalence relation. 
+ The "has the same top-level domain as" relation on a set of websites is an equivalence relation. 
+ Consider a set of triangular shapes, and put a relation on this set by saying that for any two shapes $s$ and $t$, $(s,t)$ belongs to the relation if $s$ and $t$ are [congruent](https://www.mathsisfun.com/geometry/triangles-congruent.html) as triangles. Back in 9th-grade geometry class you found out that this is an equivalence relation. 
+ Consider $\mathbb{Z}$, the set of all integers. Put a relation on $\mathbb{Z}$ by saying that for $a,b \in \mathbb{Z}$, the pair $(a,b)$ belongs to the relation if $a$ and $b$ are congruent modulo 12 -- that is, $a-b$ is a multiple of 12. This turns out to be an equivalence relation. (And it's an equivalence relation still, if you replace "12" with any positive integer.) 
+ The relation on $\{1,2,3,4\}$ shown in the directed graph below is an equivalence relation (see if you can explain why): 

```python
eqrelation = DiGraph([(1,1), (2,2), (3,3), (4,4), (2,3), (3,2), (3,4), (4,3), (2,4), (4,2)], loops=True)
eqrelation.show()
```

And here are some _non-examples_ -- that is, examples of things that are _not_ equivalence relations: 

+ Consider the "less than or equal to" relation from Lesson 10. This is _not_ an equivalence relation because although it is reflexive and transitive, it's not symmetric. 
+ Consider the "next to" relation from Lesson 10. This This is _not_ an equivalence relation because although it is reflexive and symmetric, it's not transitive. 
+ Consider the Twitter relation. This has _none_ of the properties of an equivalence relation. 

However, we could come up with some interesting equivalence relations that involve Twitter, such as: 

+ Put a relation on the set of Twitter accounts by saying that two accounts are "related" if they have the same number of followers. 
+ Or, put a relation on the set of Twitter accounts by saying that two accounts are "related" if they lost the same number of followers on February 10, 2016. 

The idea again is the __equivalence relations make precise what it means for two objects in a set to be "essentially the same" without saying they are _literally the same_.__

### Equivalence classes

Let's go back to the "rhymes with" example. A game that I used to play with my kids when they were learning about rhyming words is to _list as many words as you can that rhyme with..._ and then I'd give a word to start with. For example I might say "cow". And then the kids would start saying: 

>How, now, eyebrow, sow (_as in, a pig_), bow (_as in, bending over slightly_), ...

And so on. The game is to see how long of a list we can make (without making up nonsense words). It's especially fun when you start with a tricky word like "chicken" (thicken, quicken,...) or "orange" (um... well, "orange" rhymes with itself, but anything else?). 

Translated to our fancy math language of equivalence relations, what we are doing in this game is listing all the words that are related to a given word. What words are "related to" the word "cow"? What words are related to the word "dog"? To the word "orange"? 

We can do the same with the top-level domain example. What are some websites that are "related to" `cnn.com` in this relation? Well... lots of them. In fact anything that ends in `.com`. That's a long list! 

Giving a list -- or more precisely, a _set_ of objects that are related to a given object is an important idea, and it is encapsulated in this definition: 

>__Definition:__ Let $r$ be an equivalence relation on the set $A$, and let $x \in A$. The __equivalence class of $x$__ is the set 
$$\{ y \in A \, | \, (x,y) \in r \}$$
We denote the equivalence class of $x$ as $[x]$. 

In English, the equivalence class of $x$ is just the set of all points that are related to $x$ using the relation $r$. So we could have written: 

+ In the rhyming relation: [cow] = {cow, how, now, eyebrow, sow, bow, ...}
+ In the top level domain relation: [`cnn.com`] = {`cnn.com`, `foxnews.com`, `msnbc.com`, `amazon.com`,...}
+ In the "congruent mod 12" relation on $\mathbb{Z}$: $[4] = \{\dots, -20, -8, 4, 16, 28, 40, \dots\}$ and $[0] = \{\dots, -24, -12, 0, 12, 24, 36, \dots\}$. 
+ Also note that in the "congruent mod 12" example, $[4]$ and $[16]$ are the same set, even though they can be "represented by" different integers. 
+ In the relation above that was given as a directed graph, $[1] = \{1\}$ and $[2] = \{2,3,4\}$. _Note well the different delimiters we are using here (curly braces vs. square brackets) -- this matters._
+ Again in the directed graph example, note that $[2] = [3] = [4]$. Different "representatives" but the sets being described are literally the same. 

So one way of thinking about equivalence classes is this: __Equivalence classes are the "clusters" or "chunks" of objects that are equivalent to each other.__ I think this is especially clear in the directed graph example where the equivalent nodes literally look like clusters. As you see in the "congruent mod 12" example it's possible to have an infinite set, but only finitely many equivalence classes (there are only 12 different classes -- make sure you write these out to see for yourself.)
