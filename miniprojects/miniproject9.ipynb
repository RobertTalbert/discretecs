# Miniproject 9: The diameter of a graph 

## Overview 

This is the first of three miniprojects that explore the three main __performance metrics__ of graphs that model social networks. In this miniproject we will focus on the concept of the _diameter_ of a graph. Let's begin with thinking about what this idea describes in real-life terms. 

We know that any social network can be modeled with a graph, where the nodes in the graph are people (or users) and there is an edge between two users if they have "friended" each other -- that is, if they are connected in the network. Some social networks are best modeled by _directed_ graphs -- like Twitter, where the connections have a _direction_ to them. Others are best modeled by _undirected_ graphs -- like Facebook, where the connections are necessarily two-directional always. For this miniproject we will only consider the "two-way" social networks that are modeled by undirected graphs. 

Consider this graph that models a social network: 

```python
network1 = Graph({'Alice':['Bob'], 'Bob':['Charles'], 'Charles':['Dave'], 'Dave':['Ed']})
network1.show()
```

Now compare that to this network: 

```python
network2 = Graph({'Rob':['Steve', 'Ty', 'Ursula','Walt', 'Zeb'], 'Steve':['Ty', 'Zeb'], 'Ty':['Steve', 'Ursula'],
                  'Ursula': ['Ty', 'Walt'], 'Walt':['Ursula', 'Zeb'], 'Zeb':['Walt', 'Steve']})
network2.show()
```

Even though the second network has more people in it, you could say that it is a "tighter" or "more close-knit" network than the first one. 

What exactly is it about the first network that makes it less "tight"? Well, one thing you might notice is that in the first network, the friends are spread fairly far apart. The connection between Ed and Alice, for instance is a fairly lengthy 4 edges long -- Ed and Alice are "fourth level" friends. On the other hand, in the second network, all the friends are fairly close -- no two nodes are more than two edges away. 

This notion of _distance_ within a graph is an important feature of social networks, and we'll define two precise terms to help us study it. 

>__Definition:__ Let $u$ and $v$ be nodes in a graph. The __distance__ from $u$ to $v$ is the length of any shortest path from $u$ to $v$. We will denote this $D(u,v)$. If there is no path from $u$ to $v$, we say that $D(u,v) = \infty$. 

For example, in the first network the distance from Alice to Ed, that is D(Alice, Ed), is equal to 4. In the second network, the distance from Ty to Walt is $D(\text{Ty}, \text{Walt}) = 2$. This is because although there are lots of separate paths from Ty to Walt -- for example (Ty, Steve, Zeb, Rob, Walt) is a path of length 4 -- the _shortest_ ones are all 2 edges long. 

With that definition of distance, we can now define the main object of study for the miniproject:

>__Definition:__ Let $G$ be a graph. The __diameter__ of $G$ is the largest distance between any two nodes of the graph. The diameter is $\infty$ if there are two nodes that cannot be connected by a path. 

Remember that "distance" means the length of a _shortest_ path. So the diameter measures the length of the "longest shortest path" in the network. 

For example, the diameter of `network1` above is 4 because this is the largest distance between any two nodes. The diameter of `network2` on the other hand is just 2, because the largest distance between nodes is 2. There are some paths in the graph of length longer than 2, but the nodes connected with those paths can be connected with shorter paths, and the longest possible one of those is 2 edges long. 


So the diameter of a social network (that is, the diameter of the graph representing it) measures how closely-knit the people in the network are. Or, you can think of the diameter as a measure of how "dense" versus how "spread out" a network is. Networks with a large diameter have people in them that are connected but only through a fairly long path of connections. Networks with a smaller diameter consist of people who are very closely connected with few "high-level" connections. The complete graph $K_n$ would have diameter 1, because every pair of nodes is connected by an edge and therefore there is always a path of length 1 between any two (different) nodes.
