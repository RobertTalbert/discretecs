## Introducing Jupyter

**Jupyter**, pronounced like the planet Jupiter, is a computational environment where you can mix code, formatted text, and mathematical notation together to create rich, good-looking, and interactive documents for computing and science. At the heart of the Jupyter system is the **notebook**. Jupyter notebooks are documents where you can combine different ways of work and communication. For example you can have blocks or "cells" of executable Python code:

```python
# Returns a list of the squares from 0 to n-1
def square_list(n):
    squares = [ ]
    for i in range(n): 
        squares.append(i**2)
    return squares
```

```python
print(square_list(10))
```

```python
# For MTH 325 students: Using Python to visualize graphs...

# Jupyter has access to the huge collection of Python libraries that are available: 
import networkx as nx
import matplotlib.pyplot as plt

# Let's plot the complete graph on 20 vertices: 
k20 = nx.complete_graph(20)  
nx.draw(k20, pos = nx.circular_layout(k20), edge_color="red", 
        node_color="cyan", with_labels = True)
plt.show()
```

## Ways to get set up with Jupyter

There are lots of ways to set up a Jupyter environment that works for you. Try out one or more of the following to see what works best. 

### Method 1 -- In the cloud using Microsoft Azure

This uses the newly-created Microsoft Azure Notebook server to create and host your Juptyer notebooks. **Please note that since this is a cloud-based service, it requires internet access to use. If this is an issue for you, check out Method 2 below.**

1. Go to https://notebooks.azure.com/. 
2. Sign in using your Microsoft account (which you have through GVSU) or create a new account, then sign in. 
3. On the next page, click "Go to My Notebook Server". (Or click the other button to see examples.) 
4. Then you will see a page where you can upload or create new Jupyter notebooks. To create a new one click on the menu in the upper right and select "Python 3":

![Creating a new notebook](azure-open.png)

Now you can start editing. 

### Method 2 -- Locally, using Anaconda

This method is good for those who want to work offline. It uses **Anaconda** which is a full-featured scientific computing environment that includes Jupyter notebooks. 

1. Go to https://www.continuum.io/downloads and click on the tab for your operating system (Windows, OS X, or Linux). 
2. Use the appropriate installed for **Python 3.5**. Note that the current version number for Python might be higher than 3.5. But whatever you do *do not install Python 2.7!*
3. Follow the directions to install Anaconda on your system. 
4. When done, open up a terminal window and (at the command line) enter in the following three commands, one at a time: 
        conda update conda
        conda install jupyter
        conda install numpy scipy sympy matplotlib
This will update all the software in your installation to the latest versions. 
5. To launch the Jupyter notebook environment, stay in the terminal and, using the command line, navigate to the directory where you want the Jupyter notebooks to be stored. Then enter: `jupyter notebook`. The notebook server will then appear in a web browser window. 
6. Now you will see a page where you can upload or create new Jupyter notebooks. See Method 1, step 4. The kernel options you get may be different; choose **Python [default]** if the option is given to you. 

### Method 3 -- In the cloud using SageMath Cloud

SageMath Cloud is another cloud-based service that lets you host Jupyter notebooks along with many other kinds of documents for scientific computing. 

1. Go to http://cloud.sagemath.com and create an account, or sign in using a social media service. 
2. Create a "Project" that will house your notebooks. You can call it whatever you want; your class name would be a good default. Then click the project to open it. 
3. Click the button that has a blue plus sign on it and the word "Create". (Click the *button*, not the down-arrow attached to the button.) 

![Creating a new notebook](smc-create.png)

4. In the page that appears, type in a title for your notebook; THEN click "Jupyter Notebook". The notebook now should open, and you can begin editing. Make sure that in the **Kernel** menu, you select **Anaconda (Python 3)**. 

### Method 4 (for Mac users only) -- Locally using Pineapple

Pineapple is a macOS app currently in beta that lets you run Jupyter notebooks locally. It's very pretty and has some nice formatting options; but it is definitely a "beta" product and is not without its issues. 

To install it, just go to https://nwhitehead.github.io/pineapple/ and click on the download button, then follow the directions to install it. When you launch it, a new Juptyer notebook will open. **Make sure to download the version for Python 3.5.**

The GitHub respository for Pineapple is [here](https://github.com/nwhitehead/pineapple/), in case you want to look at the source code and follow its production.
