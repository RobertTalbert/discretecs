### Introduction

We're studying techniques for constructing sound mathematical arguments, and so far we have looked at three methods of proof: 

+ **Mathematical induction** -- Useful when proving something about a predicate over the natural numbers or positive integers. 
+ **Direct proof** -- Useful when proving a conditional or "if-then" statement. 
+ **Proof by contraposition** -- Useful when proving a conditional or "if-then" statement but assuming the hypothesis to try and prove the conclusion is awkward, and the contrapositive is simpler. 

The second two techniques apply when proving a conditional statement. In this lesson we are going to look at a third method of proof that also works when proving a conditional statement, that's a little different. 

### A courtroom drama

Let's motivate the idea by imagining the following courtroom drama:

Alice is being accused of murdering Bob on September 1 between the hours of 7pm and 8pm. She was found wandering the streets with a bloody knife in the vicinity of the crime around 8:30pm, so she looks like a prime suspect. However, her lawyer has a bombshell piece of evidence to give. 

"Your honor," says her lawyer, "My client is innocent. I will prove this to you now. Let's suppose she really did commit the crime." Gasps come from the courtroom and the jury. How could the lawyer _assume_ that his client is guilty when he is supposed to be proving she is innocent? 

He continues: "If she committed the crime, then she must have been at the scene of the crime during 7pm and 8pm. Let's suppose she was. But, I have evidence here from her employer..." Here he hands over a piece of paper to the judge. "And this evidence shows that she was at her place of employment, which is a butcher shop, until she got off work at 8:10pm. That explains the knife. But the evidence contradicts the fact that the crime took place between 7pm and 8pm. Therefore she cannot have committed the crime because she cannot be in two places at once!" 

The courtroom erupts. "Order! Order!" cries the judge as she bangs the gavel on the podium. But the lawyer slips into his seat smiling, knowing that his logic is impeccable. 

Was the lawyer's logic convincing to you? Here is the outline of his argument: 

1. Assume that Alice is not innocent. That is -- assume the exact opposite of what you want to prove. 
2. Under this assumption, she committed the murder and that means she had to be at the scene of the crime between 7pm and 8pm.
3. However, this is a contradiction, because we have incontrovertible proof that she was at work between 7pm and 8pm. 
4. Therefore the assumption that Alice is not innocent, is false. Which means that Alice _is_ innocent. 

What do you think? 

### Proof by contradiction


This method of reasoning is called _proof by contradiction_ (sometimes _indirect proof_), and it's a third way to think about the proof of a conditional statement. It's a powerful way to argue any point, and it has particular power in mathematics and CS. 

Indirect proofs (a.k.a. proofs by contradiction) have the following structure: 

1. **Assume the _negation_ of the statement you are trying to prove.** For conditional statements $p \rightarrow q$, that negation is $p \wedge \neg q$ -- that is, assume the usual hypothesis _and_ the _negation_ of the conclusion. 
2. **Using logic, previously proven results, known facts, etc., take small steps forward** to reinterpret your assumptions, just like you do in a direct proof. But, keep one eye on the assumptions that you made. 
3. **Use the assumptions against each other to arrive at a contradiction** -- something that cannot possibly be true. 
3. Because we arrived at a contradiction, the assumption that the _negation_ of our statement is true, is false. Therefore the negation of the statement is false; hence the statement itself is true. 

This is weird, so here is an example from the theory of sets to illustrate it. 

>**Proposition:** If $A$ and $B$ are sets, then $A \cap (B \setminus A) = \emptyset$. 

Again, the strategy here is to assume the _opposite_ (negation) of what you want to prove; then make small steps forward from that assumption until a contradiction is reached. At that point you can conclude that the statement to prove is true, because its negation is false. 

Here's a completed proof, broken down step by step. 

**Proof:** 
+ Suppose $A$ and $B$ are sets and the opposite of what we want to prove, that is, assume $A \cap (B \setminus A) \neq \emptyset$. (_Justification: This is the setup for a contradiction proof._)
+ Since $A \cap (B \setminus A)$ is not empty, it contains a point, which we will call $x$. (_Justification: If a set is nonempty then it has something in it._) 
+ This means $x \in A$ and $x \in B \setminus A$. (_Justification: Definition of "intersection"._) 
+ Since $x \in B \setminus A$, this means $x \in B$ and $x \not \in A$. (_Justification: Definition of "set difference"._) 
+ But now we have $x \in A$ and $x \not \in A$. This is a contradiction. (_Justification: You can't have a point both in and not in a set at the same time._)
+ Therefore $A \cap (B \setminus A) = \emptyset$. (_Justification: The negation of the statement was false because it led to a contradiction. Therefore the statement itself must be true.) 

Proof by contradiction is elegant and efficient, but it also requires that both the reader and writer keep a sharp eye out for contradictions as we move forward from the assumptions. It's not always apparent what the crucial contradiction will be from the beginning. You just have to be observant.

### Reminder about negating a conditional statement

When we use contradiction to prove a conditional statement, we begin by assuming the negation of that statement. But, what is the negation of a conditional statement? We did this in our study of logical equivalence when we showed that: 

>The negation of the conditional statement $P \rightarrow Q$ is the conjunction $P \wedge (\neg Q)$. In words, the negation of "If $P$ then $Q$" is "$P$ and not $Q$". 

**The negation of a conditional statement is not another conditional statement. It is two statements joined with an "and".** Keep this in mind during the next example. 


### Another example

Proofs by contradiction work particularly well when you are proving something "phrased in the negative" like two things being _unequal_, like we saw above. Here's another slightly more complex example: 

>__Proposition:__ For all real numbers $x,y$, if $x \neq y$, $x > 0$, and $y > 0$, then $x/y + y/x > 2$. 

This is a somewhat long statement to prove, so let's break it and its negation down. The statement to prove is: 

>If $x \neq y$, $x > 0$, and $y > 0$, then $x/y + y/x > 2$.

Therefore its negation is: 

>$x \neq y$, $x > 0$, and $y > 0$, **and** $x/y + y/x \leq 2$

Again, it's all of "$P$" (the original hypothesis) and then "not $Q$", with "and" joining them. Note that "not $Q$" involves changing the inequality $>$ to the inequality $\leq$. 

Now we can proceed with the proof. This time we will not list out the steps one by one; instead, you should do this as you read, and write out a justification for each step. If you can't do that, then either the proof is flawed because it's using an unjustifiable step or has a big gap in the reasoning, or else you might need to review some basic facts. 



__Proof:__ (By contradiction.) Assume the opposite of what we are trying to prove, namely that we have two real numbers $x$ and $y$ such that $x \neq y$, $x > 0$, and $y > 0$ and that $x/y + y/x \leq 2$. Since $x, y > 0$ we can multiply both sides of the inequality by $x$ and by $y$ without changing the direction of the inequality: 
$$(xy)(x/y) + (xy)(y/x) \leq 2xy$$
And multiplying out the left side gives 
$$x^2 + y^2 \leq 2xy$$
Now subtract over the $2xy$ to get $x^2 - 2xy + y^2 \leq 0$. 

(_Heads up: Here comes a contradiction._) However: We can factor the left side to get 
$$(x - y)^2 \leq 0$$
Since we are _squaring_ the left side, the result cannot be negative. So the only way for this inequality to be true is if $x-y = 0$. But that would mean $x=y$, which contradicts the assumption that $x \neq y$. Since we have a contradiction, the original proposition must be true. That is, it must be the case that $x/y + y/x > 2$. $\square$

What a trip! 

Notice in this proof how we used a result that we had previously proven, namely that if $n^2$ is even then $n$ is even. We proved that using contraposition. Just as how, in a well-written computer program, we develop code in a modular way and farm out helping tasks to subroutines in order to keep the main program simple, in mathematics we prove simple results off to the side and "import" them as necessary in bigger proofs whenever we need them. 

### Another example: Rational and irrational numbers 


Indirect proof is especially helpful in situations similar to proof by contraposition where there are a lot of "negative" statements being made. One very common class of statements has to do with _irrational_ versus _rational_ numbers. Here is a definition: 

>__Definition:__ A number $x$ is said to be _rational_ if there exist integers $p$ and $q$ with $q \neq 0$ and no common factors bigger than 1, such that $x = p/q$. The number $x$ is _irrational_ if it is not rational, i.e. there are no such integers. 

__Examples:__ Any valid fraction of integers is a rational number, such as $3/5$ or $-199/2$. Any finite decimal is rational because you can make it into a fraction with a power of 10 in the denominator, for example $3.14159 = 314159/100000$. Any integer is a rational number because you can write it as a fraction with $1$ on the bottom, e.g. $5 = 5/1$. However lots of numbers are irrational, some of the more famous ones being $\pi$ and $\sqrt{2}$. These cannot be represented exactly using a finite decimal or fraction. 

We use the word "rational" because it has the word "ratio" in it, which is the idea we are trying to express. One very important part of that definition is that condition that $p$ and $q$ have no common factors bigger than 1. This is in place to ensure that the fraction we form is in lowest form. For example, 10 and 12 have a common factor of 2; if you formed the fraction $10/12$, this wouldn't be in lowest form because this can be reduced to $5/6$. Another way to say that $p$ and $q$ have no common factors bigger than 1 is to say that their _greatest common divisor_ equals 1. In notation, this would be written as $\gcd(p,q) = 1$. For example $\gcd(10,12) = 2$ but $\gcd(10,9) = 1$. 

Pay attention to that condition, because it comes up big in a minute. 

We said that $\pi$ and $\sqrt{2}$ are irrational, and you may have been taught this fact in the past, but how do we really know this is the case? We would need to prove that _there exists no fraction of integers to which those numbers are equal_. But that's proving that something _doesn't exist_! How do you prove something doesn't exist? 

One answer to that question is to use contradiction. This is one of the most famous proofs in all of mathematics, and you should study and get familiar with it. 

>__Proposition:__ The number $\sqrt{2}$ is irrational. 

__Proof:__ (By contradiction.) Assume the negation of the proposition, namely that $\sqrt{2}$ is a rational number. This means that there exist integers $p,q$ with $q \neq 0$ and $\gcd(p,q) = 1$ such that 

$$\sqrt{2} = \frac{p}{q}$$

Squaring both sides gives: 
$$2 = \frac{p^2}{q^2}$$

Multiplying both sides by $q^2$ gives us $2q^2 = p^2$. This means that $p^2$ is even, since it is an integer ($q^2$) multiple of 2. Since $p^2$ is even, it follows from previous proofs that $p$ itself is even. 

Since $p$ is even, there exists an integer $k$ such that $p = 2k$. Therefore $p^2 = 4k^2$. Substituting this into $2q^2 = p^2$ gives 
$$2q^2 = 4k^2$$
Therefore $q^2 = 2k^2$, which implies that $q^2$ is even. Since $q^2$ is even, $q$ itself is even. 

We have shown at this point that both $p$ and $q$ are even. __But this is a contradiction to the assumption that $\gcd(p,q)=1$__, since the GCD of two even integers must be at least 2. 

Therefore we have arrived at a contradiction, so our supposition that $\sqrt{2}$ is rational must be false; therefore $\sqrt{2}$ is irrational. $\square$

Mind. Blown. 

You might notice a lot of similarities between proof by contradiction and proof by contraposition. In both methods of proof, if we are trying to prove $p \rightarrow q$ then we begin by assuming $\neg p$. And sometimes contraposition is more straightforward and easier to understand. However contradiction has two facets that sometimes make it preferable to contraposition:

+ Generally speaking, the more assumptions you get to make in a proof, the better -- and in a contradiction proof you get to assume not only $\neg q$ but also $p$. Two assumptions instead of one. 
+ One thing that the $\sqrt{2}$ proof shows you is that _indirect proof also works on statements that are not conditional statements_. The proposition we proved was not conditional; it just states that $\sqrt{2}$ is irrational. You can't really use contraposition unless you have an if-then statement. So indirect proof is usable on a larger set of propositions than contraposition.
